## list vs numpy

```python
py_list = [1,2,3]
print("python list multiplication", py_list * 2)

np_array = np.array([1,2,3])
print("numpy array multiplication", np_array * 2)

import time
start_time = time.time()
py_list = [i*2 for i in range(1000000)]
    
print("Time taken for python list multiplication: ", time.time() - start_time)

start_time = time.time()
np_array = np.arange(1000000) * 2
print("\n numpy operation time: ", time.time() - start_time)
```

##creating array from scratch

```python
zeros = np.zeros((3, 4))
print("zero array: \n", zeros)

ones = np.ones((2, 3))
print("one array: \n", ones)

full = np.full((2,2),7)
print("full array: \n",full)

random = np.random.random((2,3))
print("random array: \n",random)

sequence = np.arange(0, 10, 2)
print('sequance array: \n',sequence)
```

###array reshaping


```python
arr = np.arange(12)
print("origanal array", arr)

reshaped = arr.reshape((3, 4))
print("reshaped array", reshaped)

flattened =reshaped.flatten()
print("\n flattened array", flattened)

#ravel (returns view insted of copy)

raveled = reshaped.ravel()
print("\n raveled array", raveled)
```
