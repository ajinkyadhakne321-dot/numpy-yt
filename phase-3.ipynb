```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array ([
    [1, 150000, 180000, 220000, 250000], # Paradise
    [2, 120000, 140000, 160000, 190000], # Beijing
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000] # Chai Point

])

print("=== zomato sales analysis ====")
print("\n sales data shape", sales_data.shape)
print("sample data for first three restorants", sales_data[0:3])
```

```python
print("sample data for first three restorants", sales_data[:, 1:])
```

```python
###total sales per year
```

```python
print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("yearly total sales", yearly_total)
yearly_total = np.sum(sales_data[:, 1:], axis=1)
print("yearly total sales", yearly_total)
```

```python
#minimum sales per restorunt
```

```python
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)
```

```python
#maximum sales per year 
```

```python
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)
```

```python
#avg sales per restarunt
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

```python
#cumulative sales per restarunt
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print(cumulative_sales)
```

```python
plt.Figure(figsize=(10, 6))
plt.plot(sales_data[:, 0], sales_data[:, 1:], marker='o')
plt.title("Sales Trends (2021-2024)")
plt.xlabel("Restaurant ID")
plt.ylabel("Sales")
plt.xticks(sales_data[:, 0])
plt.legend(["2021", "2022", "2023", "2024"])
plt.grid()
```

#vector operations

```python
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

print("vector addition", vector1 + vector2)
print("vector subtraction", vector1 - vector2)
print("element-wise multiplication", vector1 * vector2)
print("dot product", np.dot(vector1, vector2))
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("angle between vectors (radians)", angle)
print("angle between vectors (degrees)", np.degrees(angle))
```

```python
resturant_types = np.array(["Paradise", "Beijing", "Pizza Hub", "Burger Point", "Chai Point"])
sales_2024 = sales_data[:, 4]
plt.figure(figsize=(10, 6))
plt.bar(resturant_types, sales_2024, color='skyblue')
plt.title("Sales in 2024 by Restaurant")
plt.xlabel("Restaurant")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
