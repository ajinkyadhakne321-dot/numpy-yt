```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create some sample arrays
array1 = np.array([[1, 2, 3], [5, 6, 7]])  # nested list to form a 2×3 array
array2 = np.random.random((3, 3))          # shape tuple is clearer
array3 = np.zeros((4, 4))                  # same here
```

```python
np.save("array1.npy", array1)
np.save("array2.npy", array2)
np.save("array3.npy", array3) #saving in flat files
```

```python
loaded_array1 = np.load("array1.npy")
loaded_array2 = np.load("array2.npy")
loaded_array3 = np.load("array3.npy") #loading from flat files
print("Loaded Array 1:\n", loaded_array1)
print("Loaded Array 2:\n", loaded_array2)
print("Loaded Array 3:\n", loaded_array3)
```

```python
import os
# load existing logo if available, otherwise make a simple placeholder
if os.path.exists("numpy-logo.npy"):
    logo = np.load("numpy-logo.npy")
else:
    # create a blue square as a stand-in for the numpy logo
    logo = np.zeros((64, 64, 3), dtype=np.uint8)
    logo[..., 2] = 255
plt.figure(figsize=(6, 6))
plt.imshow(logo)
plt.title("numpy logo (placeholder)")
plt.axis('off')
```
